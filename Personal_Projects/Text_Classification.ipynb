```python
#importing bert
model = TFAutoModel.from_pretrained("bert-base-uncased")
```

```python
#Raw data can't be fed into it so we tokenize the data
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
```

```python
#Example of how it tokanizes the data, token 101 and 102 represent special tokens start and end, tokens in middel represents Hi
#Attenton_mask is all 1 because there are no padding tokens

#Padding tokens: because the sent list or data might have variable sets of length

#truncation will truncate any text that has more them 512 words which is the maximum context size

#return_tensors will change a normal python list to a tensorflow tensor


inputs = tokenizer(['Prabhakar', 'How are you', 'I am fine'], padding=True, truncation=True, return_tensors='tf')
inputs
```

```python
output = model(inputs)
output
```
