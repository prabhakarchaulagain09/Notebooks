```python
from graphviz import Digraph
from IPython.display import Image

# Create a new directed graph
er = Digraph('ER Diagram', filename='er_diagram', format='png')

# Define the entities
er.node('Room', 'Room\n(RoomID, RoomType, Description, Amenities, Rate, Availability)')
er.node('Guest', 'Guest\n(GuestID, FullName, Address, PhoneNumber, Email, PaymentInfo)')
er.node('Staff', 'Staff\n(StaffID, FullName, Position, ContactNumber)')
er.node('Reservation', 'Reservation\n(ReservationID, GuestID, RoomID, StaffID, CheckInDate, CheckOutDate, Status, TotalAmount)')
er.node('RestaurantOrder', 'RestaurantOrder\n(OrderID, GuestID, ItemName, Quantity, TotalAmount, TaxAmount, Discount, GrandTotal)')
er.node('Inventory', 'Inventory\n(InventoryID, ItemName, Quantity, Location)')

# Define the relationships
er.edge('Room', 'Reservation', label='has')
er.edge('Guest', 'Reservation', label='makes')
er.edge('Staff', 'Reservation', label='creates')
er.edge('Guest', 'RestaurantOrder', label='makes')
er.edge('Reservation', 'Guest', label='for')
er.edge('Reservation', 'Staff', label='by')

# Render the diagram to a file
er.render()

# Display the image
display(Image('er_diagram.png'))
```
