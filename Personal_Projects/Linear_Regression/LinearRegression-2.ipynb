```python
x = data['SalePrice'].to_numpy()
y = data['GrLivArea'].to_numpy()

sum_x = np.sum(x)
sum_y = np.sum(y)

sum_x2 = np.sum(x**2)
sum_xy = np.sum(x*y)

n = (len(x))

m = (n*sum_xy - sum_x*sum_y)/(n*sum_x2 - (sum_x**2))
b = (sum_y*sum_x2 - sum_x*sum_xy)/(n*sum_x2 - (sum_x**2))

print('m = ',m)
print('b = ',b)
```
