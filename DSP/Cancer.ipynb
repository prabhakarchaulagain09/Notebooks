```python
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
from scipy.stats import mode, skew, kurtosis

# Load the cancer dataset
cancer = load_breast_cancer()

df = pd.DataFrame(data=cancer.data, columns=cancer.feature_names)
df['target'] = cancer.target  # Adding target column

# Display first 5 rows
df.head()
```

```python
# Dataset Information

df.info()
```

```python
# Summary statistics of numerical columns

# Extract numeric features
numeric_features = df.iloc[:, :4]

# Mean
mean_values = numeric_features.mean()

# Median
median_values = numeric_features.median()

# Mode
mode_values = numeric_features.apply(lambda x: mode(x, keepdims=True).mode[0])

# Skewness
skewness_values = numeric_features.apply(skew)

# Kurtosis
kurtosis_values = numeric_features.apply(kurtosis)

# Display results
print("\nStatistical Analysis:")
print(f"Mean:\n{mean_values}")
print(f"\nMedian:\n{median_values}")
print(f"\nMode:\n{mode_values}")
print(f"\nSkewness:\n{skewness_values}")
print(f"\nKurtosis:\n{kurtosis_values}")
```

```python
# Histogram

plt.figure(figsize=(8, 6))
plt.hist(df['mean radius'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Mean Radius')
plt.xlabel('Mean Radius')
plt.ylabel('Frequency')
plt.show()
```

```python
#Bar Garph

target_counts = df['target'].value_counts()
plt.figure(figsize=(6, 4))
plt.bar(target_counts.index, target_counts.values, color=['lightcoral', 'lightgreen'])
plt.xticks([0, 1], ['Benign', 'Malignant'])
plt.title('Frequency of Tumor Types')
plt.xlabel('Tumor Type')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot

plt.figure(figsize=(8, 6))
plt.scatter(df['mean radius'], df['mean texture'], c=df['target'], cmap='viridis', alpha=0.7)
plt.title('Mean Radius vs. Mean Texture')
plt.xlabel('Mean Radius')
plt.ylabel('Mean Texture')
plt.colorbar(label='Target (0: Benign, 1: Malignant)')
plt.show()
```

```python
#Box Plot

plt.figure(figsize=(8, 6))
sns.boxplot(x='target', y='mean area', data=df, palette=['lightgreen', 'lightcoral'])
plt.title('Distribution of Mean Area by Tumor Type')
plt.xticks([0, 1], ['Benign', 'Malignant'])
plt.xlabel('Tumor Type')
plt.ylabel('Mean Area')
plt.show()
```

```python
#Pie Chart
target_counts = df['target'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(target_counts.values, labels=['Benign', 'Malignant'], autopct='%1.1f%%', colors=['lightgreen', 'lightcoral'], startangle=90)
plt.title('Proportion of Tumor Types')
plt.show()
```

```python
# Calculating Correlation
cor = df.corr()

# Heatmap of correlations
plt.figure(figsize=(30, 30))
sns.heatmap(cor, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

# Insights from EDA on Breast Cancer Dataset


## Key Insights

1. **Tumor Type Distribution:** The dataset contains more instances of benign tumors compared to malignant tumors.


2. **Mean Radius and Texture:** There is a relationship between the mean radius and mean texture of the tumors, with malignant tumors generally having a larger mean radius and a wider range of mean texture values. This relationship is visualized in the scatter plot.


3. **Mean Area and Tumor Type:** Malignant tumors tend to have a larger mean area compared to benign tumors, as seen in the box plot. This suggests that mean area could be a useful feature for tumor classification.


4. **Feature Correlations:** There are strong correlations between several features, such as 'mean radius', 'mean perimeter', and 'mean area'. This is evident in the correlation heatmap. These features may carry similar information, and it might be possible to reduce redundancy during feature selection for model building.


5. **Data Distribution:** The histogram of 'mean radius' shows that the data is somewhat normally distributed but might have a slight positive skew. The statistical analysis of the dataset using measures like mean, median, mode, skewness, and kurtosis also supports this observation.


## Conclusion

These insights highlight potential patterns and relationships within the breast cancer dataset. They inform further analysis and model building stages, providing guidance on feature selection, model choice, and interpretation of results.
