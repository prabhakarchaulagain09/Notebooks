# Exploratory Data Analysis (EDA) on Diabetes Dataset

In this notebook, we will perform **Exploratory Data Analysis (EDA)** on the `load_diabetes` dataset from `sklearn.datasets`.

## Objectives:
- Load and understand the dataset
- Perform basic statistics & data exploration
- Visualize data distribution
- Identify correlations & patterns
- Detect and handle outliers
- Answer conceptual questions at the end

```python
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_diabetes

# Load the diabetes dataset
diabetes = load_diabetes()
df = pd.DataFrame(data=diabetes.data, columns=diabetes.feature_names)
df['target'] = diabetes.target  # Adding target column

# Display first 5 rows
df.head()
```

## 1. Basic Information and Summary Statistics

```python
# Dataset Information

df.info()
```

```python
# Summary statistics of numerical columns

```

## 2. Checking for Missing Values

```python
# Check for missing values
# Check for missing values

print("\nMissing Values:")
print(df.isnull().sum())
```

## 3. Data Visualization

```python
# Histograms for feature distributions
plt.hist(df['age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Pairplot for relationships

sns.pairplot(df)
plt.show()
```

## 4. Correlation Analysis

```python
# Calculating Correlation
cor = df.corr()

# Heatmap of correlations
plt.figure(figsize=(10, 8))
sns.heatmap(cor, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

## 5. Outlier Detection Using Boxplots

```python
# Boxplots for detecting outliers
df.columns
sns.boxplot(df[['age', 'sex', 'bmi', 'bp', 's1', 's2', 's3', 's4', 's5', 's6'] ].plot(kind='box', figsize=(10, 8)))
plt.title('Boxplots for Outlier Detection')
plt.show()
```

#6. Subplots

```python
plt.subplot(1, 2, 1)
sns.barplot(x='age', y='target', data=df)
plt.title('Age vs. Target')

plt.subplot(1, 2, 2)
sns.scatterplot(x='bmi', y='target', data=df)
plt.title('BMI vs. Target')

plt.subplot(2,2,3)
plt.hist(df["age"])

plt.subplot(2,2,4)
plt.boxplot(df["bmi"])

plt.show()
```
